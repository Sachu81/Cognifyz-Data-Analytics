# LEVEL 1 TASK 4

```python
percentage_online_delivery = df['Has Online delivery'].value_counts(normalize = True).get('Yes',0) * 100
print("percentage of restaurants that offer online delivery:" "\n",percentage_online_delivery)
```

# LEVEL 1 TASK 2 

                                  City Analysis 

```python
city_rest_count=df['City'].value_counts()

top_city = city_rest_count.idxmax()
top_city_count = city_rest_count.max()

print(f"The city with the highest number of restaurants is {top_city} with {top_city_count} restaurants.")
```

```python

city_rest_count.head(10).plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.title('Top 10 Cities with the Most Restaurants')
plt.xlabel('City')
plt.ylabel('Restaurant Count')
plt.xticks(rotation=45)
```

```python
average_ratings = df.groupby('City') ['Aggregate rating'].mean()
print(average_ratings)
```

```python
highest_avg_rating_city = average_ratings.idxmax()
highest_avg_rating = average_ratings.max()
print(f"The city with the highest average rating is {highest_avg_rating_city} with an average rating of {highest_avg_rating:.2f}.")
```

# LEVEL 2 TASK 3

```python
import plotly.express as px
```

```python
fig = px.scatter_mapbox(
    df,
    lat='Latitude',
    lon='Longitude',
    hover_name='Restaurant Name',
    zoom=2,
    height=600
)

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()


```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

coords = df[['Latitude', 'Longitude']].values

db = DBSCAN(eps=0.001, min_samples=2).fit(coords)
labels = db.labels_

df['Cluster'] = labels


plt.figure(figsize=(10, 6))
plt.scatter(df['Longitude'], df['Latitude'], c=df['Cluster'], cmap='viridis', label='Restaurant Clusters')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Clusters of Restaurants')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
split = df['Cuisines'].str.split(',').explode()
common_cuisine_combinations =split.value_counts()
common_cuisine_combinations.head(10)
```

```python
split = df['Cuisines'].str.split(',').explode()
common_cuisine_rating = pd.DataFrame({'Cuisine Combinations':split, 'Aggregate Rating': df['Aggregate rating']})
common_cuisine_rating = common_cuisine_rating.dropna()

average_cuisine_rating = common_cuisine_rating.groupby('Cuisine Combinations') ['Aggregate Rating'].mean()
top_high_rated_cuisine = average_cuisine_rating.nlargest()
print(f"top cuisine by rating :{top_high_rated_cuisine}")
```
